```python
import os
import datetime
import time


class Mcdonald:
    def __init__(self):
        orderMacdonald()




DEBUG = False
welcomeStr = '''
안녕하세요. 맥도날드입니다.
주문시간은 아래와 같습니다.

- 맥도날드 :오전 10시30분 ~ 새벽 4시
- 맥딜리버리 : 오전 10시 20분 ~ 새벽 3시50분

[MENU]

[CHICKEN]
1. 맥스파이시_타바스코_소스&나초(단품)	7600원
2. 더블_맥스파이시_상하이버거(단품)	8900원
3. 맥스파이시_상하이버거(단품) 5500원
4. 맥크리스피_클래식버거(단품)	5900원
5. 맥크리스피_디럭스버거(단품)	6800원
6. 맥치킨(단품)	3500원
7. 맥치킨 모짜렐라	5000원

[BEEF&CHEESE]
8. 쿼터파운더_치즈버거(단품)	5500원
9. 더블_쿼터파운더_치즈버거(단품)	7400원
10. 주문하기
11. 뒤로가기

메뉴를 선택해 주세요 :
'''


#메뉴 딕셔너리 생성
menuDict = {
    "맥스파이시_타바스코_소스&나초" : 7600,
    "더블_맥스파이시_상하이버거" : 8900,
    "맥스파이시_상하이버거" : 5500,
    "맥크리스피_클래식버거" : 5900,
    "맥크리스피_디럭스버거" : 6800,
    "맥치킨" : 3500,
    "맥치킨 모짜렐라" : 5000,
    "쿼터파운더_치즈버거" : 5500,
    "더블_쿼터파운더_치즈버거" : 7400
}



# 각 메뉴에 인덱스 부여하고 새로운 메뉴딕트에 넣음
menuDict2 = {}
for index,(key,value) in enumerate(menuDict.items()):
    # print(index+1,key,value)
    menuDict2[index+1] = [key,value]

if DEBUG:
    print(menuDict2)
    print(menuDict2.items())




#로그 기록용 날짜 포맷
now = datetime.datetime.now()
# now = time.localtime(time.time())
formatted_date = now.strftime('%Y-%m-%d')
formatted_time = now.strftime('%H:%M:%S')





#로그 파일 생성
def createCSV():
    csvFile = "macOrderCSV.csv"
    csvStr = "날짜,시간,메뉴명,가격\n"
    f = open(csvFile,"w")
    f.write(csvStr)
    f.close()



#로그 추가 함수
def appendCSV(orderList):
    orderList_method = orderList
    menuStr = ""
    for i in orderList_method:
        menuStr += formatted_date + "," + formatted_time + "," + menuDict2[i][0] + "," + str(menuDict2[i][1]) + "\n"
    f = open("macOrderCSV.csv","a")
    f.write(menuStr)
    f.close()



#주문하기
orderList = []
def order(orderList):
    orderList_method = orderList
    print(welcomeStr)

    while True:
        menuNum = int(input())
        if DEBUG :
            print(type(menuNum))
            print(menuNum)

        if(menuNum == 10 and orderList_method != []):
            break
        elif(menuNum == 10 and orderList_method == []):
            print("주문내역이 없습니다.")
            break
        elif(menuNum == 11):
            print(welcomeStr)
            orderList_method.clear()
        else:
            orderList_method.append(menuNum)


    ordered_items = []
    ordered_price = []
    total_price = 0
    for i in orderList:
        ordered_items.append(menuDict2[i][0])
        ordered_price.append(menuDict2[i][1])
        total_price += menuDict2[i][1]
    if orderList_method == []:
        return
    else:
        bill_ask = str(input("영수증을 드릴까요? 1.네 2.아니오"))
        if bill_ask == "1" or bill_ask == "네":
            printBill()
        else:
            print("주문을 종료합니다.")



#주문내역 저장하기
def orderSave():
    ordered_items = []
    ordered_price = []
    total_price = 0
    for i in orderList:
        ordered_items.append(menuDict2[i][0])
        ordered_price.append(menuDict2[i][1])
        total_price += menuDict2[i][1]



#로그기록
def recordLOG():
    if os.path.isfile("macOrderCSV.csv"):
        appendCSV(orderList)
    else:
        createCSV()
        appendCSV(orderList)


#영수증 출력
def printBill():
    billFile = "bill.txt"
    total = 0
    billStr = "-"*50 + "\n"
    # billStr += "%16s    |  %2s" % ("메뉴이름","가격") + "\n"
    billStr = "메뉴이름                   | 가격\n"
    billStr += "-"*50 + "\n"
    for menuNum in orderList:
        total += menuDict2[menuNum][1]
        billStr +="%16s | %4d \n" % (menuDict2[menuNum][0], menuDict2[menuNum][1])
    billStr += "-"*50 + "\n"
    billStr +="총액:{} \n".format(total)
    billStr += "-"*50 + "\n"
    print(billStr)

    f = open(billFile,"w")
    f.write(billStr)
    f.close()



# 통합함수
def orderMacdonald():
    orderList.clear()
    order(orderList)
    orderSave()
    recordLOG()



```
